```python
import numpy as np

N=1000
X = np.random.randn(N)
a = 0.3
b = 0.1
Y = a * X + b + 0.1 * np.random.randn(N)
Y_true=0.3 * X + 0.1
```

```python
import pandas as pd

df=pd.DataFrame({"X":X,"Y":Y})
print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.scatter(X,Y,alpha=0.5)
plt.plot(X,Y_true,color="red")
plt.show()
```
